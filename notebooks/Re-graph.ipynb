```python
import pickle
import matplotlib.pyplot as plt
import numpy as np
```

```python
with open('../general.pickle', 'rb') as f:
    data = pickle.load(f)
```

```python
data['gens'] = list(map(lambda x: x + 1, data['gens']))

labels = [f'Geração {x}' for x in data['gens']]
```

```python
plt.figure(figsize=(15, 8))

x = data['gens']

plt.plot(data['gens'], data['fitnesses'])
plt.title('Fitness do melhor indivíduo por geração')
plt.xticks(x, labels, rotation='vertical');
plt.ylabel('Fitness')

plt.vlines([x[0], x[-1]], 8, [data['fitnesses'][0][0], data['fitnesses'][-1][-1]], linestyles='dashdot')

plt.annotate(f'{round(data["fitnesses"][0][0])}',
            xy=(x[0], data['fitnesses'][0][0]),
            xytext=(0, 3),  # 3 points vertical offset
            textcoords="offset points",
            ha='center', va='bottom')

plt.annotate(f'{round(data["fitnesses"][-1][-1])}',
            xy=(x[-1], data['fitnesses'][-1][-1]),
            xytext=(0, 3),  # 3 points vertical offset
            textcoords="offset points",
            ha='center', va='bottom');
```

```python
with open('../a2c_params_data.pickle', 'rb') as f:
    data = pickle.load(f)
```

```python
data['gens'] = np.arange(len(data['learning_rate']))
```

```python
labels = [f'Geração {x}' for x in data['gens']]
```

```python
def rand_jitter(arr):
    stdev = .01*(max(arr)-min(arr))
    return arr + np.random.randn(len(arr)) * stdev
```

```python
y = data['pop_learning_rate'][0]
x = np.arange(1, len(y) + 1)

plt.title('Learning rate da população na primeira geração')
plt.scatter(rand_jitter(x), y)
plt.ylabel('Learning Rate');
```

```python
with open('../dqn_params_data.pickle', 'rb') as f:
    data = pickle.load(f)
```

```python
data['gens'] = np.arange(len(data['learning_rate']))
```

```python
y = data['pop_learning_rate'][0]
x = np.arange(1, len(y) + 1)

plt.title('Learning rate da população na primeira geração')
plt.scatter(rand_jitter(x), y)
plt.ylabel('Learning Rate');
```

```python
y = data['pop_learning_rate'][-1]
x = np.arange(1, len(y) + 1)

plt.title('Learning rate da população na última geração')
plt.scatter(rand_jitter(x), y)
plt.ylabel('Learning Rate');
```

```python
data['pop_gamma'] = data['pop_gamma'][:-1]
```

```python
y = data['pop_gamma'][0]
x = np.arange(1, len(y) + 1)

plt.title('Discount Factor da população na primeira geração')
plt.scatter(rand_jitter(x), y)
plt.ylabel('Discount Factor');
```

```python
y = data['pop_gamma'][-1]
x = np.arange(1, len(y) + 1)

plt.title('Discount Factor da população na última geração')
plt.scatter(rand_jitter(x), rand_jitter(y))
plt.ylabel('Discount Factor');
```
