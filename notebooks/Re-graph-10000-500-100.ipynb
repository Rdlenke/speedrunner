```python
import pickle
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = {'machines': [1, 2, 3, 4, 5], 'time': [2441, 1191, 921, 761, 642]}

plt.figure(figsize=(8, 5))

plt.title('Tempo de treinamento conforme quantidade de máquinas')
x = np.arange(1, 6)
plt.plot(data['machines'], data['time'])
plt.yticks(np.arange(750, 3000, 250))
plt.xticks(np.arange(1, 6))
plt.xlabel('Máquinas')
plt.ylabel('Tempo em minutos');

for index in range(len(data['time'])):
    if index == 1:
        xytext = (10, 7)
    else:
        xytext = (0, 3)
    plt.annotate(f'{data["time"][index]}',
                xy=(x[index], data["time"][index]),
                xytext=xytext,  # 3 points vertical offset
                textcoords="offset points",
                ha='center', va='bottom');
```

```python
with open('../graph-10000-500-100/a2c_params_data.pickle', 'rb') as f:
    data = pickle.load(f)
```

```python
data['gens'] = np.arange(0, 100)
```

```python
data['gens'] = [x + 1 if index == 0 else x + 2 for index, x in enumerate(data['gens'])]

labels = [f'Geração {x}' for x in data['gens']]
```

```python
def rand_jitter(arr):
    stdev = .01*(max(arr)-min(arr))
    return arr + np.random.randn(len(arr)) * stdev
```

```python
y = data['pop_learning_rate'][0]
x = np.arange(1, len(y) + 1)

plt.title('Learning rate da população na primeira geração')
plt.scatter(rand_jitter(x), y)
plt.ylabel('Learning Rate');
```

```python
y = data['pop_learning_rate'][-1]
x = np.arange(1, len(y) + 1)

plt.title('Learning rate da população na última geração')
plt.scatter(rand_jitter(x), y)
plt.ylabel('Learning Rate');
```

```python
X = np.arange(len(data['pop_learning_rate']))
y = [np.std(x) for x in data['pop_learning_rate']]


labels = [f'Geração {x + 1}' for x in X]

plt.figure(figsize=(25, 8))

plt.title('Desvio padrão do learning rate conforme as gerações')
plt.plot(X, y)
plt.xticks(X, labels, rotation='vertical');
```

```python
y = data['pop_gamma'][0]
x = np.arange(1, len(y) + 1)

plt.title('Discount Factor da população na primeira geração')
plt.scatter(rand_jitter(x), y)
plt.ylabel('Discount Factor');
```

```python
y = data['pop_gamma'][-1]
x = np.arange(1, len(y) + 1)

plt.title('Discount Factor da população na última geração')
plt.scatter(rand_jitter(x), y)
plt.ylabel('Discount Factor');
```

```python
X = np.arange(len(data['pop_gamma']))
y = [np.std(x) for x in data['pop_gamma']]


labels = [f'Geração {x + 1}' for x in X]

plt.figure(figsize=(25, 8))

plt.title('Desvio padrão do discount factor conforme as gerações')
plt.plot(X, y)
plt.xticks(X, labels, rotation='vertical');
```
